```python
# Google Colabで実行する場合に必要なライブラリをインストールします
# 既にインストール済みでも、セッションがリセットされる場合に備えて実行しておくのが安全です
!pip install chromedriver-autoinstaller
!pip install selenium
!pip install pandas
!apt-get update
!apt-get install -y google-chrome-stable

import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
import glob
import os

# ユーザー情報
myid = 'okada@azm.co.jp'
mypw = 'artesiA0902'

# URL
login_url = 'https://mitsumo-rich.jp/'
project_url = 'https://mitsumo-rich.jp/project/'

# ダウンロード先のパスを設定
download_path = '/content/downloads'

# Chromeオプションを設定
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# ダウンロード設定
prefs = {
    'download.default_directory': download_path,
    'download.prompt_for_download': False,
    'download.directory_upgrade': True,
    'safebrowsing.enabled': True
}
options.add_experimental_option('prefs', prefs)

# WebDriverを起動
chromedriver_autoinstaller.install()
driver = webdriver.Chrome(options=options)

try:
    # ログイン
    print("ログイン中...")
    driver.get(login_url)
    wait = WebDriverWait(driver, 20)  # ログイン処理のタイムアウトを20秒に延長

    id_field = wait.until(EC.presence_of_element_located((By.NAME, "id")))
    pw_field = driver.find_element(By.NAME, "pw")
    id_field.send_keys(myid)
    pw_field.send_keys(mypw)
    login_button = driver.find_element(By.CSS_SELECTOR, 'input[type="submit"][value="ログイン"]')
    driver.execute_script("arguments[0].click();", login_button)

    try:
        wait.until(EC.url_to_be("https://mitsumo-rich.jp/"))
        print("ログインに成功しました。")
    except:
        print("ログイン後のページ遷移に失敗しました。")
        try:
            continue_login_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//input[@value='ログインを続ける']")))
            continue_login_button.click()
            print("「ログインを続ける」ボタンをクリックしました。")
            wait.until(EC.url_to_be("https://mitsumo-rich.jp/"))
        except Exception as e:
            print(f"「ログインを続ける」ボタンが見つからなかったか、クリックできませんでした。エラー: {e}")
            raise

    '''
    # ログイン後のページに遷移したことを確認するための待機
    # ログイン後のページに表示される要素（例: 「案件管理」メニューなど）を待機
    print("ログイン後のページロードを待機しています...")
    try:
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'a[href="/project/"]')))
        print("ログイン成功。")
    except:
        print("ログイン後のページへの遷移に失敗したか、要素が見つかりませんでした。")
        raise

    # 「ログインを続ける」ボタンが表示されたらクリックする処理
    try:
        # ボタンのCSSセレクタはウェブサイトによって異なるため、必要に応じて修正してください
        keep_login_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#btn_continue_login')))
        keep_login_button.click()
        print("「ログインを続ける」ボタンをクリックしました。")
        # ボタンクリック後のページ遷移を待機
        time.sleep(3)
    except:
        print("「ログインを続ける」ボタンは表示されませんでした。")
        pass
    '''

    print("案件管理ページへアクセスします。")
    driver.get(project_url)
    wait.until(EC.title_contains("案件管理"))

    # 「詳細検索」リンクをクリックし、フォームを表示
    detail_search_link = wait.until(EC.element_to_be_clickable((By.ID, "show_detail_search")))
    detail_search_link.click()
    print("「詳細検索」をクリックしました。")
    wait.until(EC.visibility_of_element_located((By.ID, "detail_search")))
    print("詳細検索フォームが表示されました。")

    # 検索条件を入力
    print("検索条件を入力中...")
    unselected_values = ["1", "3", "4"]
    for val in unselected_values:
        checkbox = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'input[name="status_kn[]"][value="{val}"]')))
        if checkbox.is_selected():
            driver.execute_script("arguments[0].click();", checkbox)
            print(f"チェックボックス (value={val}) をクリックしました。")

    status_checkbox = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'input[name="status_kn[]"][value="2"]')))
    if not status_checkbox.is_selected():
        driver.execute_script("arguments[0].click();", status_checkbox)
        print("「受注」のチェックボックスをクリックしました。")

    # 検索ボタンをクリック
    search_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div#detail_search input[type="submit"][value="検索"]')))
    driver.execute_script("arguments[0].click();", search_button)
    print("検索を実行しました。")

    # 検索結果のテーブルが読み込まれるのを待つ
    wait.until(EC.presence_of_element_located((By.ID, "project_list")))
    time.sleep(2)

    # ダウンロード先のディレクトリを作成
    os.makedirs(download_path, exist_ok=True)

    # CSVダウンロードボタンをJavaScriptでクリック
    download_button = driver.find_element(By.ID, "download")
    driver.execute_script("arguments[0].click();", download_button)
    print("CSVダウンロードを開始しました。")

    # ダウンロードが完了するまで待機
    timeout = 30
    start_time = time.time()
    while not glob.glob(os.path.join(download_path, '*.csv')) and time.time() - start_time < timeout:
        time.sleep(1)

    csv_files = glob.glob(os.path.join(download_path, '*.csv'))
    if not csv_files:
        raise FileNotFoundError("CSVファイルが見つかりませんでした。")

    csv_file_path = csv_files[0]
    print(f"CSVファイルのパス: {csv_file_path}")

    # ----------------------------------------
    # CSVファイルをpandasのデータフレームに読み込む
    # ----------------------------------------
    try:
        # まずはUTF-8で試行
        df = pd.read_csv(csv_file_path, encoding='utf-8')
        print("CSVファイルをUTF-8で正常に読み込みました。")
    except UnicodeDecodeError:
        try:
            # UTF-8で失敗した場合、cp932で再試行
            df = pd.read_csv(csv_file_path, encoding='cp932')
            print("CSVファイルをcp932で正常に読み込みました。")
        except Exception as e:
            # cp932でも失敗した場合
            print(f"CSVファイルの読み込みに失敗しました。原因: {e}")
            df = None

    if df is not None:
        print("\n--- データフレームの最初の5行 ---")
        print(df.head())
        print("---------------------------------")

finally:
    driver.quit()
```
